```python
# libraries!
import numpy as np      # numpy is Python's "array" library
import pandas as pd     # Pandas is Python's "data" library ("dataframe" == spreadsheet)
```

```python
# let's read in our flower data...
# 
# for read_csv, use header=0 when row 0 is a header row
# 
filename = 'titanic.csv'
df = pd.read_csv(filename)        # encoding="utf-8" et al.
print(f"{filename} : file read into a pandas dataframe.")
```

```python
#
# a dataframe is a "spreadsheet in Python"   (seems to have an extra column!)
#
# let's view it!
df
```

```python
# Let's look at the dataframe's columns:
df.columns
```

```python
# we can drop a series of data (a row or a column)
# the dimensions each have a numeric value, row~0, col~1, but let's use readable names we define:
ROW = 0
COLUMN = 1

df_clean1 = df.drop('the original Kaggle dataset!   Adapted from here:   https://www.kaggle.com/c/titanic', axis=COLUMN)
df_clean1
```

```python
# we can drop a series of data (a row or a column)
# the dimensions each have a numeric value, row~0, col~1, but let's use readable names we define:
ROW = 0
COLUMN = 1

df_clean2 = df_clean1.drop('boat', axis=COLUMN)
df_clean2
```

```python
# we can drop a series of data (a row or a column)
# the dimensions each have a numeric value, row~0, col~1, but let's use readable names we define:
ROW = 0
COLUMN = 1

df_clean3 = df_clean2.drop('body', axis=COLUMN)

df_clean4 = df_clean3.drop('fare', axis=COLUMN)

df_clean5 = df_clean4.drop('ticket', axis=COLUMN)

df_clean6 = df_clean5.drop('cabin', axis=COLUMN)

df_clean7 = df_clean6.drop('home.dest', axis=COLUMN)

df_clean8 = df_clean7.drop('embarked', axis=COLUMN)

df_clean9 = df_clean8.drop('name', axis=COLUMN)
df_clean9
```

```python
# let's drop _all_ rows with data that is missing/NaN (not-a-number)
df_clean10 = df_clean9.dropna()  # drop na rows (NaN, not-a-number)
df_clean10
```

```python
df_clean10.info()   # Get the information of the data leftover
```

```python
COLUMNS = df_clean10.columns            # "list" of columns
print(f"COLUMNS is {COLUMNS}")  
  # It's a "pandas" list, called an Index
  # use it just as a Python list of strings:
print(f"COLUMNS[0] is {COLUMNS[0]}\n")

# let's create a dictionary to look up any column index by name
COL_INDEX = {}
for i, name in enumerate(COLUMNS):
    COL_INDEX[name] = i  # using the name (as key), look up the value (i)
print(f"COL_INDEX is {COL_INDEX}")
print(f"COL_INDEX[ 'survived' ] is {COL_INDEX[ 'survived' ]}")
```

```python
# First, let's map our below and above median strings to numeric values:

SURVIVAL = ['deceased', 'survived']   # int to str
SURVIVAL_INDEX = {'survived':1,'deceased':0}  # str to int

def convert_survival(survival):
    """ return the survival value (a unique integer) """
    return SURVIVAL_INDEX[survival]

GENDER = ['female', 'male']
GENDER_INDEX = {'female': 0, 'male': 1}

def convert_gender(gender):
    """ return the gender value (a unique integer) """
    return GENDER_INDEX[gender]

# Let's try it out...
for name in SURVIVAL:
    print(f"{name} maps to {convert_survival(name)}")

for gender in GENDER:
    print(f"{gender} maps to {convert_gender(gender)}")
```

```python
#
# we can "apply" to a whole column and create a new column
#   it may give a warning, but this is ok...
#

df_clean11 = df_clean10.copy()  # copy everything AND...

# edit the above/below median column
df_clean11['sex'] = df_clean10['sex'].apply(convert_gender)

# let's see...
pd.options.display.min_rows = 30
df_clean11
```

```python
#
# we can "apply" to a whole column and create a new column
#   it may give a warning, but this is ok...
#

df_clean12 = df_clean11.copy()  # copy everything

# let's see...
pd.options.display.min_rows = 30
df_clean12
```

```python
# Let's call our final table as df_tidy...
df_tidy =  df_clean12
```

```python
# We'll construct the new filename:
old_basename = filename[:-4]                      # remove the ".csv"
cleaned_filename = old_basename + "_cleaned.csv"  # name-creating
print(f"cleaned_filename is {cleaned_filename}")

# Now, save
df_tidy.to_csv(cleaned_filename, index_label=False)  # no "index" column...
```
